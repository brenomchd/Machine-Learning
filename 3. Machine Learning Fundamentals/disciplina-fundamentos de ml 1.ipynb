# Exercício 1:

Faça um programa que leia duas listas de 10 posições (utilizando a função input) e
construa outra lista contendo, nas posições pares, os valores da primeira e, nas posições
ímpares, os valores da segunda.

```python
lista1 = [int(input(f"Digite o {i+1}º número da lista1: ")) for i in range(10)]
lista2 = [int(input(f"Digite o {i+1}º número da lista2: ")) for i in range(10)]

lista_final = []

for i in range(10):
    lista_final.append(lista1[i])   # valores da lista1
    lista_final.append(lista2[i])   # valores da lista2

print("Lista final:", lista_final)
```

# Exercício 2

Faça um programa que crie uma lista de 10 elementos aleatórios, calcule e imprima:**negrito**

Exemplo:

1.   A média dos elementos;
2.   O maior e menor elemento;
3.   A quantidade de elementos positivos e a quantidade de elementos
negativos;
4.   Quantos elementos estão abaixo da média e quantos estão acima;
5.   Desvio padrão;

```python
import random
import statistics

# Gerando 10 números aleatórios entre -50 e 50
lista = [random.randint(-50, 50) for _ in range(10)]
print("Lista:", lista)

# a) média
media = statistics.mean(lista)

# b) maior e menor
maior = max(lista)
menor = min(lista)

# c) positivos e negativos
positivos = len([x for x in lista if x > 0])
negativos = len([x for x in lista if x < 0])

# d) abaixo/acima da média
abaixo = len([x for x in lista if x < media])
acima = len([x for x in lista if x > media])

# e) desvio padrão
desvio = statistics.pstdev(lista)

print(f"Média: {media}")
print(f"Maior: {maior}, Menor: {menor}")
print(f"Positivos: {positivos}, Negativos: {negativos}")
print(f"Abaixo da média: {abaixo}, Acima da média: {acima}")
print(f"Desvio padrão: {desvio}")
```

# Exercício 6

Faça uma função que, dada uma tupla com N notas, retorne a média das notas.

```python
def media_notas(notas: tuple):
    return sum(notas) / len(notas)

print(media_notas((7, 8, 9, 10)))
```

# Exercício 7

Escreva uma função que receba uma lista de elementos e retorne a quantidade de
elementos únicos (distintos) na lista.

```python
def elementos_unicos(lista):
    return len(set(lista))

print(elementos_unicos([1,2,2,3,4,4,5]))
```

# Exercício 8

Escreva uma função que receba uma lista de elementos e retorne a quantidade de
elementos duplicados na lista.

```python
def elementos_duplicados(lista):
    return len(lista) - len(set(lista))

print(elementos_duplicados([1,2,2,3,4,4,5]))
```
