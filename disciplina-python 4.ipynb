# Questão 2

```python
import random, statistics

lista = [random.randint(-50, 50) for _ in range(10)]
print("Lista:", lista)

media = sum(lista) / len(lista)
print("Média:", media)

print("Maior elemento:", max(lista))
print("Menor elemento:", min(lista))

positivos = sum(1 for x in lista if x > 0)
negativos = sum(1 for x in lista if x < 0)
print("Positivos:", positivos, "Negativos:", negativos)

abaixo = sum(1 for x in lista if x < media)
acima = sum(1 for x in lista if x > media)
print("Abaixo da média:", abaixo, "Acima da média:", acima)

print("Desvio padrão:", statistics.pstdev(lista))
```

# Questão 6

```python
def media_notas(notas: tuple):
    return sum(notas) / len(notas) if notas else 0

print(media_notas((7, 8, 9, 10)))
```

# Questão 7

```python
def contar_unicos(lista):
    return len(set(lista))

print(contar_unicos([1, 2, 2, 3, 4, 4, 5]))
```

# Questão 8

```python
def contar_duplicados(lista):
    vistos = set()
    duplicados = set()
    for elem in lista:
      if elem in vistos:
          duplicados.add(elem)
      else:
        vistos.add(elem)
    return len(duplicados)

print(contar_duplicados([1, 2, 2, 3, 4, 4, 5]))
```
