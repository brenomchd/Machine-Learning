# Exercício 23

Faça uma pequena análise de dados com o NumPy
No site https://www.mockaroo.com/ , gere um arquivo CSV com os seguintes
parâmetros:

Captura de tela 2025-10-02 214226.png

Faça um script em Python para ler o arquivo CSV para uma matriz NumPy e Calcule
e imprima:

1. Média, maior e menor idade
2. Média, maior e menor idade por sexo
3. Qual sexo tem maior média e maior idade?
4. Quantas pessoas estão acima da média de idade do grupo todo e por sexo?

```python
import numpy as np
import pandas as pd

file_path = "data.csv"
df = pd.read_csv(file_path)

df.columns = ["id", "primeiro_nome", "ultimo_nome", "sexo", "idade"]

idades = df["idade"].to_numpy()
sexos = df["sexo"].to_numpy()

# 1. Média, maior e menor idade
media_idade = np.mean(idades)
maior_idade = np.max(idades)
menor_idade = np.min(idades)

print("1. Geral:")
print(f"   Média: {media_idade:.2f}, Maior: {maior_idade}, Menor: {menor_idade}\n")

# 2. Média, maior e menor idade por sexo
masc_idades = idades[sexos == "M"]
fem_idades = idades[sexos == "F"]

media_masc = np.mean(masc_idades)
media_fem = np.mean(fem_idades)

max_masc = np.max(masc_idades)
max_fem = np.max(fem_idades)

min_masc = np.min(masc_idades)
min_fem = np.min(fem_idades)

print("2. Por sexo:")
print(f"   Masculino -> Média: {media_masc:.2f}, Máx: {max_masc}, Mín: {min_masc}")
print(f"   Feminino  -> Média: {media_fem:.2f}, Máx: {max_fem}, Mín: {min_fem}\n")

# 3. Qual sexo tem maior média e maior idade?
sexo_maior_media = "Masculino" if media_masc > media_fem else "Feminino"
sexo_maior_idade = "Masculino" if max_masc > max_fem else "Feminino"

print("3. Comparação entre sexos:")
print(f"   Sexo com maior média de idade: {sexo_maior_media}")
print(f"   Sexo com maior idade máxima: {sexo_maior_idade}\n")

# 4. Quantos estão acima da média (geral e por sexo)?
acima_media_total = np.sum(idades > media_idade)
acima_media_masc = np.sum(masc_idades > media_masc)
acima_media_fem = np.sum(fem_idades > media_fem)

print("4. Pessoas acima da média:")
print(f"   Grupo todo: {acima_media_total}")
print(f"   Homens acima da média masculina: {acima_media_masc}")
print(f"   Mulheres acima da média feminina: {acima_media_fem}")
```
