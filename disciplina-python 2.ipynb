# Exercício 1

Reescreva o programa para calcular o pagamento de um funcionário (Atividade 1),
considerando que o empregado receberá um bônus de 50% por hora trabalhada
além das 40 horas obrigatórias.

Exemplo:

1.   Informe a quantidade de horas trabalhadas: 45
2.   Informe o valor da hora: 10
3.   Total a receber: 475,00

```python
x = int(input("Informe a quantidade de horas trabalhadas:"))
y = int(input("Informe o valor da hora:"))

if x <= 40:
  total = x*y

else:
  w = 40
  z = x - 40
  total = (w*y) + (z*y*1.5)
print("Total a receber: ", total)
```

# Exercício 2

Refaça o que se pede no exercício 1. Desta vez, crie e utilize uma função
chamada `calculapagamento` que recebe dois parâmetros (horas e taxa).

```python
def calculapagamento (horas, taxa):

  if horas <= 40:
    total = horas*taxa
  else:
    horas_normais = 40
    horas_extras = horas - 40
    total = (horas_normais*taxa) + (horas_extras*taxa*1.5)
  return total

horas = int(input("Informe a quantidade de horas trabalhadas:"))
taxa = int(input("Informe o valor da hora:"))

pagamento = (calculapagamento(horas, taxa))
print("Total a receber: ", pagamento)
```

# Exercício 3

Crie uma lista definida manualmente com 10 elementos, e faça um código para
encontrar qual o menor elemento armazenado na lista.

```python
menor = None

print('Antes', menor)

for numero in [2 , 56, 70, 36, 15, 58, 10, 27, 99, 47]:
  if menor is None:
    menor = numero
  elif numero < menor:
    menor = numero
  print(menor, numero)

print('Depois', menor)
```

# Exercício 4

Crie uma lista definida manualmente com 10 elementos inteiros. Faça um
código para buscar um elemento na lista, e mostre o seu funcionamento
considerando os dois cenários possíveis:

1.   Busca de um elemento que pertence à lista
2.   Busca de um elemento que não pertence à lista



```python
lista = [2 , 56, 70, 36, 15, 58, 10, 27, 99, 47]

busca = int(input("Digite um número para buscar: "))

if busca in lista:
    print(f"O número {busca} está na lista.")
else:
    print(f"O número {busca} não está na lista.")
```
