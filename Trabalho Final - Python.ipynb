# **ANÁLISE DE DADOS**

```python
df = pd.read_csv("dados.csv")
```

```python
df.columns = ["ID", "Nome", "Sobrenome", "Genero", "Idade"]

idades = df["Idade"].to_numpy()
generos = df["Genero"].to_numpy()
```

```python
# 1) Média, maior e menor idade
media_idade = np.mean(idades)
maior_idade = np.max(idades)
menor_idade = np.min(idades)
print("Média:", media_idade)
print("Maior idade:", maior_idade)
print("Menor idade:", menor_idade)
```

```python
# 2) Média, maior e menor idade por gênero
for g in np.unique(generos):
  subset = idades[generos == g]
  print(f"Gênero {g}:")
  print(" Média:", np.mean(subset))
  print(" Maior:", np.max(subset))
  print(" Menor:", np.min(subset))
```

```python
# 3) Qual sexo tem maior média e maior idade?
medias = {g: np.mean(idades[generos == g]) for g in np.unique(generos)}
maiores = {g: np.max(idades[generos == g]) for g in np.unique(generos)}
print("Maior média de idade:", max(medias, key=medias.get))
print("Maior idade registrada:", max(maiores, key=maiores.get))
```

```python
# 4) Quantas pessoas estão acima da média geral?
acima_media_geral = np.sum(idades > media_idade)
print("Acima da média geral:", acima_media_geral)
```

```python
# 5) Quantas pessoas estão acima da média em cada gênero?
for g in np.unique(generos):
  subset = idades[generos == g]
  print(f"{g}: {np.sum(subset > np.mean(subset))} pessoas acima da média do próprio gênero")
```

```python
# 6) Histograma das idades
plt.hist(idades, bins=10, color="skyblue", edgecolor="black")
plt.title("Histograma de Idades")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()
```

```python
# 7) Gráfico de dispersão das idades
plt.scatter(range(len(idades)), idades, c="red")
plt.title("Dispersão das Idades")
plt.xlabel("Índice")
plt.ylabel("Idade")
plt.show()
```
