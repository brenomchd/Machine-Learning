# Questão 6

```python
try:
  score = float(input("Digite a Pontuação (0.0 a 1.0): "))
  if score < 0.0 or score > 1.0:
    print("Pontuação Inválida")
  elif score >= 0.9:
    print("Nota equivalente: A")
  elif score >= 0.8:
    print("Nota equivalente: B")
  elif score >= 0.7:
    print("Nota equivalente: C")
  elif score >= 0.6:
    print("Nota equivalente: D")
  else:
    print("Nota equivalente: F")
except:
  print("Pontuação Inválida")
```

# Questão 8

```python
def computarNotas(score):
  if score < 0.0 or score > 1.0:
    return "Pontuação Inválida"
  elif score >= 0.9:
    return "A"
  elif score >= 0.8:
    return "B"
  elif score >= 0.7:
    return "C"
  elif score >= 0.6:
    return "D"
  else:
    return "F"


try:
  score = float(input("Digite a Pontuação (0.0 a 1.0): "))
  print("Nota equivalente:", computarNotas(score))
except:
  print("Pontuação Inválida")
```

# Questão 9

```python
soma = 0
qtd = 0

while True:
  num = int(input("Digite um número (0 para sair): "))
  if num == 0:
    break

soma += num
qtd += 1

if qtd > 0:
  media = soma / qtd

print("Soma:", soma)
print("Quantidade de números:", qtd)
print("Média:", media)
```

# Questão 10

```python
soma = 0
qtd = 0

while True:
  num = int(input("Digite um número (0 para sair): "))
  if num == 0:
      break
  if num < 0:
      print("Entrada inválida")
      continue
  soma += num
  qtd += 1

if qtd > 0:
  media = soma / qtd

print("Soma:", soma)
print("Quantidade de números:", qtd)
print("Média:", media)
```

# Questão 11

```python
soma = 0
qtd = 0
valores = []

while True:
  num = int(input("Digite um número (0 para sair): "))
  if num == 0:
      break
  soma += num
  qtd += 1
  valores.append(num)

if qtd > 0:
  print("Soma:", soma)
  print("Quantidade de números:", qtd)
  print("Máximo:", max(valores))
  print("Mínimo:", min(valores))
```
